```python
import random

def valuation_engine(input_data):
    # Base inputs
    circle_rate = input_data.get("circle_rate", 5000)  # ₹ per sq ft
    size = input_data["size"]  # sq ft
    age = input_data["age"]
    location_score = input_data.get("location_score", 0.7)
    legal_clear = input_data.get("legal_clear", True)
    rental_yield = input_data.get("rental_yield", 0.03)

    # --- Market Value ---
    base_value = circle_rate * size

    location_multiplier = 1 + (location_score - 0.5)

    # Depreciation
    if age < 5:
        age_factor = 1.0
    elif age < 15:
        age_factor = 0.9
    else:
        age_factor = 0.75

    rental_boost = 1 + rental_yield

    market_value = base_value * location_multiplier * age_factor * rental_boost

    # Range (±10%)
    market_range = [int(market_value * 0.9), int(market_value * 1.1)]

    # --- Liquidity Score ---
    resale_index = int(
        50 +
        (location_score * 30) +
        (10 if legal_clear else -15) +
        (10 if age < 10 else -10)
    )

    resale_index = max(0, min(100, resale_index))

    # --- Distress Value ---
    liquidity_discount = 0.2 if resale_index > 70 else 0.35
    distress_value = market_value * (1 - liquidity_discount)
    distress_range = [int(distress_value * 0.9), int(distress_value * 1.1)]

    # --- Time to Sell ---
    if resale_index > 80:
        time_range = [30, 60]
    elif resale_index > 50:
        time_range = [60, 120]
    else:
        time_range = [120, 240]

    # --- Confidence ---
    confidence = round(random.uniform(0.6, 0.85), 2)

    # --- Output ---
    return {
        "market_value_range": market_range,
        "distress_value_range": distress_range,
        "resale_potential_index": resale_index,
        "estimated_time_to_sell_days": time_range,
        "confidence_score": confidence,
        "key_drivers": [
            "location_score",
            "property_age",
            "rental_yield"
        ],
        "risk_flags": [
            "low_liquidity" if resale_index < 50 else "none"
        ]
    }


# Example usage
input_data = {
    "circle_rate": 6000,
    "size": 1200,
    "age": 8,
    "location_score": 0.75,
    "legal_clear": True,
    "rental_yield": 0.04
}

result = valuation_engine(input_data)
print(result)
```
