### Adaptive Quiz Game
This code implements a simple adaptive quiz. The quiz gets harder if the user answers correctly and easier if they answer incorrectly. It uses a nested dictionary to store questions grouped by difficulty.

```python
questions = {
    'easy': [
        {
            'question': 'What is 2 + 2?',
            'options': ['3', '4', '5'],
            'answer': '4'
        },

        {
            'question': 'Which planet is known as the Red Planet?',
            'options': ['Earth', 'Mars', 'Jupiter'],
            'answer': 'Mars'
        }
    ],
    'medium': [
        {
            'question': 'What is the capital of France?',
            'options': ['Berlin', 'Madrid', 'Paris'],
            'answer': 'Paris'
        },
        {
            'question': 'What is the chemical symbol for water?',
            'options': ['O2', 'H2O', 'CO2'],
            'answer': 'H2O'
        }
    ],
    'hard': [
        {
            'question': 'What is the square root of 144?',
            'options': ['10', '12', '14'],
            'answer': '12'
        },
        {
            'question': 'Who painted the Mona Lisa?',
            'options': ['Vincent van Gogh', 'Pablo Picasso', 'Leonardo da Vinci'],
            'answer': 'Leonardo da Vinci'
        }
    ]
}

def run_quiz():
    current_difficulty = 'easy'
    score = 0
    max_questions_per_difficulty = 2
    difficulty_levels = ['easy', 'medium', 'hard']
    current_level_index = 0

    print("Welcome to the Adaptive Quiz!")
    print("Answer questions to adjust the difficulty.")

    question_counts = {level: 0 for level in difficulty_levels}

    while True:
        if question_counts[current_difficulty] >= len(questions[current_difficulty]):
            print(f"You've exhausted all questions at the {current_difficulty} level. Moving on if possible or ending quiz.")
            if current_level_index + 1 < len(difficulty_levels):
                current_level_index += 1
                current_difficulty = difficulty_levels[current_level_index]
                print(f"Moving to {current_difficulty} difficulty.")
            else:
                print("No more questions available.")
                break

        available_questions = [q for q in questions[current_difficulty] if not q.get('asked', False)]

        if not available_questions:
            print(f"All questions for {current_difficulty} difficulty have been asked. Moving to next difficulty.")
            if current_level_index + 1 < len(difficulty_levels):
                current_level_index += 1
                current_difficulty = difficulty_levels[current_level_index]
                print(f"Moving to {current_difficulty} difficulty.")
                continue # Try to get a question from the new difficulty
            else:
                print("No more questions available across all difficulties.")
                break

        import random
        question_data = random.choice(available_questions)
        question_data['asked'] = True # Mark question as asked

        print(f"\n--- Difficulty: {current_difficulty.capitalize()} ---")
        print(question_data['question'])
        for i, option in enumerate(question_data['options']):
            print(f"{i+1}. {option}")

        user_answer = input("Your answer (enter the number): ")

        try:
            user_choice_index = int(user_answer) - 1
            if 0 <= user_choice_index < len(question_data['options']):
                selected_option = question_data['options'][user_choice_index]
                if selected_option == question_data['answer']:
                    print("Correct!")
                    score += 1
                    # Increase difficulty if not already at the hardest level
                    if current_level_index + 1 < len(difficulty_levels):
                        current_level_index += 1
                        current_difficulty = difficulty_levels[current_level_index]
                        print(f"Difficulty increased to {current_difficulty}.")
                else:
                    print(f"Incorrect. The answer was {question_data['answer']}.")
                    # Decrease difficulty if not already at the easiest level
                    if current_level_index - 1 >= 0:
                        current_level_index -= 1
                        current_difficulty = difficulty_levels[current_level_index]
                        print(f"Difficulty decreased to {current_difficulty}.")
            else:
                print("Invalid option number. Try again.")
                continue # Ask the same question again due to invalid input
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue # Ask the same question again due to invalid input

        play_again = input("Continue quiz? (yes/no): ").lower()
        if play_again != 'yes':
            break

    print(f"\nQuiz over! Your final score: {score}")

# To run the quiz, uncomment the line below:
run_quiz()
```
