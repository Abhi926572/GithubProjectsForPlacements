```python
import sys
import re

PERSONAL_PROVIDERS = {
    "gmail.com", "yahoo.com", "outlook.com", "hotmail.com",
    "icloud.com", "proton.me", "protonmail.com", "aol.com",
    "live.com", "mail.com"
}

TLD_REGION = {
    "com": "global", "net": "global", "org": "global",
    "info": "global", "io": "global", "ai": "global",
    "dev": "global", "app": "global",
    "in": "India", "us": "United States", "uk": "United Kingdom",
    "de": "Germany", "fr": "France", "jp": "Japan", "cn": "China",
    "au": "Australia", "ca": "Canada", "br": "Brazil", "za": "South Africa"
}

email_re = re.compile(r"^[^@\s]+@[^@\s]+\.[^@\s]+$")


def classify_email(email: str):
    email = email.strip()
    if not email_re.match(email):
        return "INVALID", "UNKNOWN", "UNKNOWN"

    username, domain = email.split("@", 1)
    domain = domain.lower()
    provider = domain

    if domain in PERSONAL_PROVIDERS:
        kind = "personal"
    else:
        kind = "work"

    tld = domain.split(".")[-1]
    region = TLD_REGION.get(tld, "UNKNOWN")

    return kind, provider, region


def main():
    data = sys.stdin.read().strip().splitlines()
    if not data:
        return

    t = int(data[0].strip())
    emails = data[1:1 + t]

    for email in emails:
        username, domain = ("", "")
        if email_re.match(email.strip()):
            username, domain = email.split("@", 1)
        kind, provider, region = classify_email(email)
        print(f"{username} {domain} {kind} {provider} {region}")


if __name__ == "__main__":
    main()
```
